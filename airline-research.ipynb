```
# prompt: import data from the google drive

from google.colab import drive
# # New section
drive.mount('/content/drive')
# # New section
drive.mount('/content/drive') # Mount to the drive root directory

# Now you can access your file at:
file_path = '/content/drive/My Drive/file_path/twitterdataset/airlineSentiment.csv'
# Use this file_path to read your data

import pandas as pd
df = pd.read_csv(file_path, encoding='latin-1')
```

```
# prompt: create a pie-chart of all the airline in dataset

import matplotlib.pyplot as plt

airline_counts = df['airline'].value_counts()

# Creating the pie chart
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Airlines in Dataset')
plt.show()
```

```
# prompt: create a bar graph of each airline with positive, negative and neutral airline_sentiment

import matplotlib.pyplot as plt
sentiment_counts = df.groupby(['airline', 'airline_sentiment']).size().unstack()

sentiment_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Sentiment Distribution by Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

```
# prompt: create a pie-chart of united airline with positive negative and neutral airline_sentiment

import matplotlib.pyplot as plt
# Filter the DataFrame for United Airlines only
united_df = df[df['airline'] == 'United']

# Count the sentiments for United Airlines
united_sentiment_counts = united_df['airline_sentiment'].value_counts()

# Creating the pie chart for United Airlines
plt.pie(united_sentiment_counts, labels=united_sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Sentiment Distribution for United Airlines')
plt.show()
```

```
# prompt: create a pie-chart of united airline with positive negative and neutral airline_sentiment

import matplotlib.pyplot as plt
# Filter the DataFrame for United Airlines only
united_df = df[df['airline'] == 'Virgin America']

# Count the sentiments for United Airlines
united_sentiment_counts = united_df['airline_sentiment'].value_counts()

# Creating the pie chart for United Airlines
plt.pie(united_sentiment_counts, labels=united_sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Sentiment Distribution for Virgin Airlines')
plt.show()
```

```
# prompt: create bar graph of airline_sentiment with positive, negative and neutral

import matplotlib.pyplot as plt
sentiment_counts = df['airline_sentiment'].value_counts()

plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Overall Sentiment Distribution')
plt.show()
```

```
# prompt: create bar graph of negativereson

import matplotlib.pyplot as plt
negative_reason_counts = df['negativereason'].value_counts()

plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(negative_reason_counts.index, negative_reason_counts.values)
plt.xlabel('Negative Reason')
plt.ylabel('Count')
plt.title('Distribution of Negative Reasons')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
